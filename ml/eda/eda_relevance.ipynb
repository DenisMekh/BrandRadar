```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/clean/relevance_v2.csv')
```

```python
df.shape
```

```python
print(df["relevant"].value_counts())
print(df["relevant"].value_counts(normalize=True) * 100)

plt.figure(figsize=(6,4))
sns.countplot(data=df, x="relevant")
plt.title("Relevant distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, y="company",
              order=df["company"].value_counts().index)
plt.title("Company distribution")
plt.tight_layout()
plt.show()
```

```python
# Длина текста
df["char_len"] = df["text"].str.len()
df["word_count"] = df["text"].apply(lambda x: len(x.split()))

print(df[["char_len", "word_count"]].describe())

plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="relevant", y="word_count")
plt.title("Word count by relevant")
```
